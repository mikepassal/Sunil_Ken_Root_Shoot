```python
def get_and_save_gene(gene_name):
    try:
        shoot_gene = shoot_network[[gene_name]]
        shoot_gene_sorted = shoot_gene.sort_values(by=gene_name, ascending=False)
        shoot_gene_sorted.to_csv(
            f"/data/passala/Generated_Tables/Sunil_root_shoot_project/Interesting_gene_coexpression_lists_round_2/{gene_name}_shoot_sorted_coexpression_partners.csv"
            )
    except:
        print(f'Gene {gene_name} not found in shoot network')


    try:
        root_gene = root_network[[gene_name]]
        root_gene_sorted = root_gene.sort_values(by=gene_name, ascending=False)
        root_gene_sorted.to_csv(
            f"/data/passala/Generated_Tables/Sunil_root_shoot_project/Interesting_gene_coexpression_lists_round_2/{gene_name}_root_sorted_coexpression_partners.csv"
            )
    except: 
        print(f'Gene {gene_name} not found in root network')





```

```python
gene_list = ['Zm00001eb156910','Zm00001eb360250','Zm00001eb363810','Zm00001eb062820','Zm00001eb049070','Zm00001eb207010']
```

```python
for gene in gene_list:
    get_and_save_gene(gene)
```
